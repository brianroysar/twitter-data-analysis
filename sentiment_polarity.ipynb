```python
df = pd.read_csv("/content/gdrive/MyDrive/Winter 23/6.1/vader_results.csv")
```

```python
df = df.drop(columns=['word_scores', "pos",	"neu", 	"neg",	"but_count", "Unnamed: 0"])
```

```python
grouped_df = df.groupby('user_location2').describe()
```

```python
df_final = grouped_df[grouped_df['compound']["count"] > 25]  
```

```python
df_final.columns = df_final.columns.to_flat_index()
```

```python
df_final = df_final.rename(columns={("compound", "count"): "count", ("compound", "min"): "min",("compound", "max"): "max",})
df_final = df_final.rename(columns={("compound", "mean"): "mean", ("compound", "std"): "std",("compound", "25%"): "25%",})
df_final = df_final.rename(columns={("compound", "50%"): "50%", ("compound", "75%"): "75%",})
df_final
```

```python
df_final["range"] = df_final["max"] - df_final["min"]
df_final
```

```python
df_final = df_final.sort_values(by=['range'])
df_final
```

```python
#Dropping the invalid rows
df_final = df_final.drop(["Earth", "ùüèùüñ+ ùêåùêÉùêçùêà", "Inside your router"])
df_final
```

```python
df_final = df_final.sort_values(by=['std'])
data_head = df_final.head(3)
data_tail = df_final.tail(3)
df_concat = pd.concat([data_head, data_tail])
df_concat
```

```python
mins = df_concat["min"]
maxes = df_concat["max"]
means = df_concat["mean"]
std = df_concat["std"]
regions = list(df_concat.index.values)

plt.subplots(figsize=(18,5))
plt.errorbar(regions, means, std, fmt='ok', lw=3)
plt.errorbar(regions, means, [means - mins, maxes - means],
             fmt='.k', ecolor='gray', lw=1)
plt.xlabel("Region")
plt.ylabel("Sentiment")
plt.title("Tweet Sentiment Spread per Region")
plt.xlim(-1, len(regions))
```
